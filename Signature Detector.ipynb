```python
"""
Object Detection (On Image) From TF2 Saved Model
=====================================
"""
import os
import glob
# import img2pdf
import tensorflow as tf
import cv2
import time
from object_detection.utils import label_map_util, config_util
from object_detection.utils import visualization_utils as viz_utils
import argparse
from google.colab.patches import cv2_imshow


# PROVIDE PATH TO IMAGE DIRECTORY
IMAGE_PATHS = path

# PROVIDE PATH TO MODEL DIRECTORY
PATH_TO_MODEL_DIR = '/content/drive/MyDrive/PDF-Redaction/Tensorflow/workspace/training_demo/exported-models/my_model'

# PROVIDE PATH TO LABEL MAP
PATH_TO_LABELS = '/content/drive/MyDrive/PDF-Redaction/Tensorflow/workspace/training_demo/annotations/label_map.pbtxt'

# PROVIDE THE MINIMUM CONFIDENCE THRESHOLD
MIN_CONF_THRESH = float(0.50)




print('Loading model...', end='')
start_time = time.time()
# LOAD THE MODEL
PATH_TO_SAVED_MODEL = PATH_TO_MODEL_DIR + "/saved_model"

# LOAD SAVED MODEL AND BUILD DETECTION FUNCTION
detect_fn = tf.saved_model.load(PATH_TO_SAVED_MODEL)

end_time = time.time()
elapsed_time = end_time - start_time
print('Done! Took {} seconds'.format(elapsed_time))

# LOAD LABEL MAP DATA FOR PLOTTING
category_index = label_map_util.create_category_index_from_labelmap(PATH_TO_LABELS,
                                                                    use_display_name=True)

import numpy as np
# from PIL import Image
# import matplotlib.pyplot as plt
# import warnings
# warnings.filterwarnings('ignore')   # Suppress Matplotlib warnings

# def load_image_into_numpy_array(path):
#     """Load an image from file into a numpy array.
#     Puts image into numpy array to feed into tensorflow graph.
#     Note that by convention we put it into a numpy array with shape
#     (height, width, channels), where channels=3 for RGB.
#     Args:
#       path: the file path to the image
#     Returns:
#       uint8 numpy array with shape (img_height, img_width, 3)
#     """
#     return np.array(Image.open(path))


for img in IMAGE_PATHS:
  print('Running inference for {}... '.format(img), end='')

  image = cv2.imread(img)
  image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
  image_expanded = np.expand_dims(image_rgb, axis=0)

  # The input needs to be a tensor, convert it using `tf.convert_to_tensor`.
  input_tensor = tf.convert_to_tensor(image_rgb)
  # The model expects a batch of images, so add an axis with `tf.newaxis`.
  input_tensor = input_tensor[tf.newaxis, ...]

  # input_tensor = np.expand_dims(image_np, 0)
  detections = detect_fn(input_tensor)

  # All outputs are batches tensors.
  # Convert to numpy arrays, and take index [0] to remove the batch dimension.
  # We're only interested in the first num_detections.
  num_detections = int(detections.pop('num_detections'))
  detections = {key: value[0, :num_detections].numpy()
                for key, value in detections.items()}
  detections['num_detections'] = num_detections

  # detection_classes should be ints.
  detections['detection_classes'] = detections['detection_classes'].astype(np.int64)

  image_with_detections = image.copy()

  # # SET MIN_SCORE_THRESH BASED ON YOU MINIMUM THRESHOLD FOR DETECTIONS
  # viz_utils.visualize_boxes_and_labels_on_image_array(
  #       image_with_detections,
  #       detections['detection_boxes'],
  #       detections['detection_classes'],
  #       detections['detection_scores'],
  #       category_index,
  #       use_normalized_coordinates=True,
  #       max_boxes_to_draw=5,
  #       min_score_thresh=0.2,
  #       agnostic_mode=False)
  
  detection_threshold=.4 #confidence of detection
  height = image.shape[0]
  width = image.shape[1]
  box_index = 0

  for detection in detections['detection_boxes']:
        ymin = int(detection[0]*height)
        xmin = int(detection[1]*width)
        ymax = int(detection[2]*height)
        xmax = int(detection[3]*width)
        if detections['detection_scores'][box_index] > detection_threshold:
          cv2.rectangle(image_with_detections, (xmin, ymin), (xmax, ymax), (0, 0, 0), -1)
        
        box_index += 1
  cv2_imshow(image_with_detections)
  cv2.imwrite(f'/content/drive/MyDrive/PDF-Redaction/Tensorflow/inference-results/{os.path.basename(img)}', image_with_detections)
  # cv2.imwrite(f'/content/modified/{os.path.basename(img)}', image_with_detections)


  # with open("redacted.pdf","wb") as f:
	#   f.write(img2pdf.convert(glob.glob("/content/modified/*.jpg")))

# print("Done modifying")
  

    # print('Done')
    # DISPLAYS OUTPUT IMAGE
    # cv2.imwrite(f'/content/drive/MyDrive/PDF-Redaction/Tensorflow/inference-results/{os.path.basename(img)}', image_with_detections)
  
```

```python
import spacy
from pdf2image import convert_from_path
# import PIL
from IPython.display import display, Image
from PIL import ImageDraw
import easyocr
import numpy as np
import cv2
```

```python
reader = easyocr.Reader(['en'], gpu=True)
```

```python
Date_extractor = spacy.load("en_core_web_sm")
```

```python
from transformers import AutoTokenizer, AutoModelForTokenClassification
from transformers import pipeline

tokenizer = AutoTokenizer.from_pretrained("dslim/bert-base-NER")
model = AutoModelForTokenClassification.from_pretrained("dslim/bert-base-NER")
```

```python
Name_extractor = pipeline("ner", model=model, tokenizer=tokenizer)
```

```python
out = reader.readtext('SWZ_C_NERCHA_SignedGrantConfirmation_1.pdf_4.jpg')
```

```python
%%time
def detect_name(image, bbox):
    im = cv2.imread(image)
    for bound in bbox:
        x1, y1 = [min(id) for id in zip(*bound[0])]
        x2, y2 = [max(id) for id in zip(*bound[0])]
        date_entity = Date_extractor(bound[1])
        try:
            name_entity = Name_extractor(bound[1])
            if (name_entity[0]['entity']=="B-PER") or (name_entity[0]['entity']=="I-PER"):
                cv2.rectangle(im, (int(x1), int(y1)), (int(x2), int(y2)), (0,0,0), -1)
        except IndexError as e:
            pass
        
        for entity in date_entity.ents:
            if (entity.label_ == "CARDINAL") or entity.label_ == "DATE":
                cv2.rectangle(im, (int(x1), int(y1)), (int(x2), int(y2)), (0,0,0), -1)
         
            
            
    return im
    
im = detect_name('SWZ_C_NERCHA_SignedGrantConfirmation_1.pdf_4.jpg',out)
```
